````python
import random
import json
import os
from pathlib import Path

import numpy as np

SEED = 42
random.seed(SEED)
np.random.seed(SEED)

ROOT = Path('/home/mk/CarCV')
DATA_DIR = ROOT / 'data/external/ymad_cars'
IMG_DIR  = DATA_DIR / 'images'
INDEX    = DATA_DIR / 'images_index.jsonl'

MODEL_BASELINE = ROOT / 'models/vehiclemakenet_pruned_onnx_v1.1.0/resnet18_pruned.onnx'
MODEL_FINETUNED = ROOT / 'models/vmn_vmmrdb_ymad_105c/model.onnx'
LABELS_BASELINE  = ROOT / 'models/vehiclemakenet_pruned_onnx_v1.1.0/labels.txt'
LABELS_FINETUNED = ROOT / 'models/vmn_vmmrdb_ymad_105c/labels.txt'

FIG_DIR = ROOT / 'notebooks/vehiclemakenet_eval'
FIG_DIR.mkdir(exist_ok=True)

RESULTS_DIR = ROOT / 'results/vehiclemakenet_eval'
RESULTS_DIR.mkdir(parents=True, exist_ok=True)

BATCH_SIZE = 128
IMG_SIZE   = (224, 224)  # VehicleMakeNet input

print('ROOT:', ROOT)
print('IMG_DIR:', IMG_DIR, '| exists:', IMG_DIR.exists())
print('MODEL_BASELINE:', MODEL_BASELINE.exists())
print('MODEL_FINETUNED:', MODEL_FINETUNED.exists())
````

## 1. Загрузка данных

````python
def load_labels(path: Path) -> list[str]:
    return path.read_text().strip().split(';')

labels_baseline  = load_labels(LABELS_BASELINE)
labels_finetuned = load_labels(LABELS_FINETUNED)

print(f'Baseline labels  ({len(labels_baseline)}): {labels_baseline}')
print(f'Finetuned labels ({len(labels_finetuned)}): {labels_finetuned[:10]} ...')
````

````python
records = []
with open(INDEX, 'r', encoding='utf-8') as f:
    seen = set()
    for line in f:
        r = json.loads(line)
        key = r['image_id']
        if key in seen:
            continue
        seen.add(key)
        img_path = IMG_DIR / r['image_path'].split('/')[-1]
        if img_path.exists():
            records.append({'path': str(img_path), 'brand': r['brand']})

print(f'Total unique images on disk: {len(records)}')

brand_counts = {}
for r in records:
    brand_counts[r['brand']] = brand_counts.get(r['brand'], 0) + 1
print(f'Unique brands: {len(brand_counts)}')
print('Top-10 brands:', sorted(brand_counts.items(), key=lambda x: -x[1])[:10])
````

````python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import seaborn as sns

set_baseline  = set(labels_baseline)
set_finetuned = set(labels_finetuned)

records_b = [r for r in records if r['brand'] in set_baseline]
records_f = [r for r in records if r['brand'] in set_finetuned]

print(f'Images valid for baseline  (20 cls): {len(records_b)} '
      f'| brands: {len({r["brand"] for r in records_b})}')
print(f'Images valid for finetuned (105 cls): {len(records_f)} '
      f'| brands: {len({r["brand"] for r in records_f})}')
print(f'Out-of-vocab brands (neither model covers):', 
      sorted(set(brand_counts) - set_finetuned))

fig, axes = plt.subplots(1, 2, figsize=(18, 6))
for ax, recs, title in [
    (axes[0], records_b, f'Baseline ({len(set_baseline)} classes)'),
    (axes[1], records_f, f'Fine-tuned ({len(set_finetuned)} classes)'),
]:
    bc = {}
    for r in recs:
        bc[r['brand']] = bc.get(r['brand'], 0) + 1
    brands_sorted = sorted(bc.items(), key=lambda x: -x[1])
    brands, counts = zip(*brands_sorted)
    ax.barh(brands, counts, color='steelblue')
    ax.set_xlabel('Images')
    ax.set_title(title)
    ax.invert_yaxis()
plt.tight_layout()
plt.savefig(FIG_DIR / 'class_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved class_distribution.png')
````

## 2. Загрузка моделей (ONNX Runtime)

````python
import onnxruntime as ort

print('ONNX Runtime version:', ort.__version__)
print('Available providers:', ort.get_available_providers())

PROVIDERS = ['CUDAExecutionProvider', 'CPUExecutionProvider']

sess_b = ort.InferenceSession(str(MODEL_BASELINE),  providers=PROVIDERS)
sess_f = ort.InferenceSession(str(MODEL_FINETUNED), providers=PROVIDERS)

def session_info(sess, name):
    inp = sess.get_inputs()[0]
    out = sess.get_outputs()[0]
    print(f'{name}: input={inp.name} {inp.shape}, output={out.name} {out.shape}')
    print(f'  Provider: {sess.get_providers()[0]}')

session_info(sess_b, 'Baseline')
session_info(sess_f, 'Finetuned')
````

## 3. Инференс

````python
import cv2
import time
from collections import defaultdict

def preprocess_batch(paths: list[str], size: tuple[int, int] = (224, 224)) -> np.ndarray:
    """
    VehicleMakeNet preprocessing:
    - Resize to 224x224
    - BGR -> RGB
    - Normalize to [0, 1]
    - NCHW float32
    """
    imgs = []
    for p in paths:
        img = cv2.imread(p)
        if img is None:
            img = np.zeros((*size, 3), dtype=np.uint8)
        img = cv2.resize(img, size, interpolation=cv2.INTER_LINEAR)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        imgs.append(img)
    batch = np.stack(imgs).astype(np.float32) / 255.0
    return batch.transpose(0, 3, 1, 2)  # NCHW


def run_inference(
    sess: ort.InferenceSession,
    records: list[dict],
    labels: list[str],
    batch_size: int = 128,
) -> dict:
    label2idx = {l: i for i, l in enumerate(labels)}
    inp_name  = sess.get_inputs()[0].name
    out_name  = sess.get_outputs()[0].name

    all_gt_idx, all_top1, all_top3_preds, all_confs = [], [], [], []
    t0 = time.perf_counter()

    for i in range(0, len(records), batch_size):
        batch_recs = records[i : i + batch_size]
        paths = [r['path'] for r in batch_recs]
        gts   = [r['brand'] for r in batch_recs]

        tensor = preprocess_batch(paths)
        logits = sess.run([out_name], {inp_name: tensor})[0]  # (B, C)
        probs  = logits  # already softmax or logits - treat as scores

        top3_idx  = np.argsort(probs, axis=1)[:, ::-1][:, :3]
        top1_idx  = top3_idx[:, 0]
        conf      = probs[np.arange(len(batch_recs)), top1_idx]

        for j, gt in enumerate(gts):
            gt_idx = label2idx.get(gt, -1)
            all_gt_idx.append(gt_idx)
            all_top1.append(top1_idx[j])
            all_top3_preds.append(top3_idx[j].tolist())
            all_confs.append(float(conf[j]))

        if (i // batch_size) % 10 == 0:
            pct = 100 * (i + len(batch_recs)) / len(records)
            print(f'  [{pct:5.1f}%] batch {i//batch_size} / {len(records)//batch_size}')

    elapsed = time.perf_counter() - t0
    fps = len(records) / elapsed
    print(f'Done: {len(records)} images in {elapsed:.1f}s ({fps:.0f} fps)')

    return {
        'gt_idx':      np.array(all_gt_idx),
        'top1_idx':    np.array(all_top1),
        'top3_preds':  all_top3_preds,
        'confidences': np.array(all_confs),
        'labels':      labels,
        'records':     records,
        'elapsed':     elapsed,
        'fps':         fps,
    }
````

````python
print('=== Baseline (20 classes) ===')
results_b = run_inference(sess_b, records_b, labels_baseline, BATCH_SIZE)

print('\n=== Fine-tuned (105 classes) ===')
results_f = run_inference(sess_f, records_f, labels_finetuned, BATCH_SIZE)
````

## 4. Метрики

````python
from sklearn.metrics import (
    accuracy_score, top_k_accuracy_score,
    classification_report, confusion_matrix,
)

def compute_metrics(res: dict, name: str) -> dict:
    gt      = res['gt_idx']
    top1    = res['top1_idx']
    top3    = res['top3_preds']
    labels  = res['labels']
    n_cls   = len(labels)

    # Фильтруем записи с валидным gt_idx (-1 = не в лейбле)
    valid   = gt >= 0
    gt_v    = gt[valid]
    top1_v  = top1[valid]
    top3_v  = [top3[i] for i in range(len(top3)) if valid[i]]
    confs_v = res['confidences'][valid]

    top1_acc = accuracy_score(gt_v, top1_v)
    top3_corr = sum(1 for g, preds in zip(gt_v, top3_v) if g in preds)
    top3_acc  = top3_corr / len(gt_v)

    # Per-class F1
    present_labels = sorted(set(gt_v.tolist()))
    report = classification_report(
        gt_v, top1_v,
        labels=present_labels,
        target_names=[labels[i] for i in present_labels],
        output_dict=True,
        zero_division=0,
    )

    metrics = {
        'name':      name,
        'n_images':  int(valid.sum()),
        'n_classes': len(present_labels),
        'top1_acc':  round(top1_acc, 4),
        'top3_acc':  round(top3_acc, 4),
        'macro_f1':  round(report['macro avg']['f1-score'], 4),
        'weighted_f1': round(report['weighted avg']['f1-score'], 4),
        'fps':       round(res['fps'], 1),
    }

    print(f'\n--- {name} ---')
    for k, v in metrics.items():
        print(f'  {k}: {v}')

    return {
        **metrics,
        'report': report,
        'gt_v': gt_v,
        'top1_v': top1_v,
        'confs_v': confs_v,
        'present_labels': present_labels,
    }


metrics_b = compute_metrics(results_b, 'Baseline-20')
metrics_f = compute_metrics(results_f, 'Finetuned-105')
````

````python
import pandas as pd

TARGET_TOP1 = 0.90

df_summary = pd.DataFrame([
    {k: v for k, v in m.items() if k not in ('report', 'gt_v', 'top1_v', 'confs_v', 'present_labels')}
    for m in [metrics_b, metrics_f]
])
df_summary['top1_pass'] = df_summary['top1_acc'] >= TARGET_TOP1
df_summary['top3_pass'] = df_summary['top3_acc'] >= TARGET_TOP1
print(df_summary.to_string(index=False))
````

## 5. Визуализации

````python
# --- 5.1 Confusion Matrix ---

def plot_confusion_matrix(m: dict, path: Path, max_classes: int = 30):
    gt_v   = m['gt_v']
    top1_v = m['top1_v']
    labels = results_b['labels'] if m['name'].startswith('Baseline') else results_f['labels']
    pl     = m['present_labels']

    if len(pl) > max_classes:
        # Только top-N по кол-ву изображений
        counts = {idx: (gt_v == idx).sum() for idx in pl}
        pl = sorted(counts, key=lambda x: -counts[x])[:max_classes]
        mask = np.isin(gt_v, pl)
        gt_v, top1_v = gt_v[mask], top1_v[mask]

    class_names = [labels[i] for i in pl]
    cm = confusion_matrix(gt_v, top1_v, labels=pl)
    cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True).clip(min=1)

    fig, axes = plt.subplots(1, 2, figsize=(max(12, len(pl) * 0.5), max(10, len(pl) * 0.45)))
    for ax, data, title, fmt in [
        (axes[0], cm,      'Raw counts',  'd'),
        (axes[1], cm_norm, 'Normalized',  '.2f'),
    ]:
        sns.heatmap(
            data, ax=ax,
            xticklabels=class_names, yticklabels=class_names,
            cmap='Blues', fmt=fmt, annot=len(pl) <= 20,
            linewidths=0.3 if len(pl) <= 20 else 0,
        )
        ax.set_title(f'{m["name"]} — Confusion Matrix ({title})')
        ax.set_xlabel('Predicted')
        ax.set_ylabel('True')
        ax.tick_params(axis='x', rotation=90)
        ax.tick_params(axis='y', rotation=0)

    plt.tight_layout()
    plt.savefig(path, dpi=150, bbox_inches='tight')
    plt.show()
    print(f'Saved {path.name}')


plot_confusion_matrix(metrics_b, FIG_DIR / 'cm_baseline.png')
plot_confusion_matrix(metrics_f, FIG_DIR / 'cm_finetuned.png', max_classes=30)
````

````python
# --- 5.2 Per-class F1 bar chart ---

def plot_per_class_f1(m: dict, res: dict, path: Path, top_n: int = 40):
    labels  = res['labels']
    report  = m['report']
    pl      = m['present_labels']

    rows = []
    for idx in pl:
        cls = labels[idx]
        if cls in report:
            r = report[cls]
            rows.append({'class': cls, 'precision': r['precision'], 'recall': r['recall'], 'f1': r['f1-score'], 'support': r['support']})

    df = pd.DataFrame(rows).sort_values('f1')
    if len(df) > top_n:
        df = pd.concat([df.head(top_n // 2), df.tail(top_n // 2)])

    fig, ax = plt.subplots(figsize=(10, max(8, len(df) * 0.35)))
    y = range(len(df))
    ax.barh(y, df['f1'],       height=0.6, label='F1',        color='#2196F3', alpha=0.9)
    ax.barh(y, df['precision'],height=0.3, label='Precision',  color='#4CAF50', alpha=0.7)
    ax.barh(y, df['recall'],   height=0.3, label='Recall',     color='#FF9800', alpha=0.7, left=0)
    ax.set_yticks(list(y))
    ax.set_yticklabels(df['class'].tolist(), fontsize=9)
    ax.axvline(x=0.9, color='red', linestyle='--', linewidth=1, label='Target 0.90')
    ax.set_xlabel('Score')
    ax.set_title(f'{m["name"]} — Per-class P/R/F1 (top/bottom {top_n // 2} by F1)')
    ax.legend(loc='lower right')
    plt.tight_layout()
    plt.savefig(path, dpi=150, bbox_inches='tight')
    plt.show()
    print(f'Saved {path.name}')


plot_per_class_f1(metrics_b, results_b, FIG_DIR / 'per_class_baseline.png')
plot_per_class_f1(metrics_f, results_f, FIG_DIR / 'per_class_finetuned.png')
````

````python
# --- 5.3 Confidence distribution: correct vs incorrect ---

def plot_confidence_dist(m: dict, path: Path):
    gt_v    = m['gt_v']
    top1_v  = m['top1_v']
    confs_v = m['confs_v']

    correct_mask = gt_v == top1_v
    c_correct = confs_v[correct_mask]
    c_wrong   = confs_v[~correct_mask]

    fig, ax = plt.subplots(figsize=(8, 5))
    bins = np.linspace(confs_v.min(), confs_v.max(), 50)
    ax.hist(c_correct, bins=bins, alpha=0.7, label=f'Correct (n={len(c_correct)})', color='green')
    ax.hist(c_wrong,   bins=bins, alpha=0.7, label=f'Wrong (n={len(c_wrong)})',   color='red')
    ax.set_xlabel('Confidence (max logit / score)')
    ax.set_ylabel('Count')
    ax.set_title(f'{m["name"]} — Confidence Distribution')
    ax.legend()
    plt.tight_layout()
    plt.savefig(path, dpi=150, bbox_inches='tight')
    plt.show()
    print(f'Saved {path.name}')


plot_confidence_dist(metrics_b, FIG_DIR / 'confidence_baseline.png')
plot_confidence_dist(metrics_f, FIG_DIR / 'confidence_finetuned.png')
````

````python
# --- 5.4 Summary bar chart vs targets ---

fig, ax = plt.subplots(figsize=(9, 5))
x = np.arange(4)
w = 0.3
metric_keys = ['top1_acc', 'top3_acc', 'macro_f1', 'weighted_f1']
metric_labels = ['Top-1 Acc', 'Top-3 Acc', 'Macro F1', 'Weighted F1']

for i, (m, color) in enumerate([(metrics_b, '#2196F3'), (metrics_f, '#FF5722')]):
    vals = [m[k] for k in metric_keys]
    ax.bar(x + i * w - w/2, vals, w, label=m['name'], color=color, alpha=0.85)
    for xi, v in zip(x + i * w - w/2, vals):
        ax.text(xi, v + 0.005, f'{v:.3f}', ha='center', va='bottom', fontsize=8)

ax.axhline(y=TARGET_TOP1, color='red', linestyle='--', linewidth=1.5, label=f'Target {TARGET_TOP1}')
ax.set_xticks(x)
ax.set_xticklabels(metric_labels)
ax.set_ylim(0, 1.1)
ax.set_ylabel('Score')
ax.set_title('VehicleMakeNet — Metrics Summary (Baseline vs Fine-tuned)')
ax.legend()
plt.tight_layout()
plt.savefig(FIG_DIR / 'metrics_summary.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved metrics_summary.png')
````

## 6. Анализ ошибок (Error Analysis)

````python
# --- 6.1 Worst-case: высокая уверенность при неправильном предсказании ---

def show_error_examples(m: dict, res: dict, path: Path, n: int = 12):
    gt_v    = m['gt_v']
    top1_v  = m['top1_v']
    confs_v = m['confs_v']
    records = res['records']
    labels  = res['labels']

    valid_mask = res['gt_idx'] >= 0
    recs_valid = [r for r, v in zip(records, valid_mask) if v]

    wrong_mask  = gt_v != top1_v
    wrong_idxs  = np.where(wrong_mask)[0]
    if len(wrong_idxs) == 0:
        print('No errors!')
        return

    # Топ-N с наивысшей уверенностью
    worst = sorted(wrong_idxs, key=lambda i: -confs_v[i])[:n]

    cols = 4
    rows_n = (len(worst) + cols - 1) // cols
    fig, axes = plt.subplots(rows_n, cols, figsize=(cols * 3.5, rows_n * 3.2))
    axes = axes.flatten()

    for ax_i, idx in enumerate(worst):
        img = cv2.imread(recs_valid[idx]['path'])
        if img is None:
            continue
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (224, 224))
        axes[ax_i].imshow(img)
        gt_lbl   = labels[gt_v[idx]]
        pred_lbl = labels[top1_v[idx]]
        axes[ax_i].set_title(
            f'GT: {gt_lbl}\nPred: {pred_lbl}\nConf: {confs_v[idx]:.3f}',
            fontsize=8, color='red'
        )
        axes[ax_i].axis('off')

    for ax_i in range(len(worst), len(axes)):
        axes[ax_i].axis('off')

    plt.suptitle(f'{m["name"]} — Worst-case Errors (high-confidence mistakes)', fontsize=11)
    plt.tight_layout()
    plt.savefig(path, dpi=120, bbox_inches='tight')
    plt.show()
    print(f'Saved {path.name}')


show_error_examples(metrics_b, results_b, FIG_DIR / 'errors_baseline.png')
show_error_examples(metrics_f, results_f, FIG_DIR / 'errors_finetuned.png')
````

````python
# --- 6.2 Per-class error rate: процент ошибок по классам ---

def print_per_class_error_rate(m: dict, res: dict, top_n: int = 15):
    gt_v    = m['gt_v']
    top1_v  = m['top1_v']
    labels  = res['labels']
    pl      = m['present_labels']

    rows = []
    for idx in pl:
        total = (gt_v == idx).sum()
        wrong = ((gt_v == idx) & (top1_v != idx)).sum()
        if total > 0:
            rows.append({'class': labels[idx], 'total': total, 'wrong': wrong, 'error_rate': wrong / total})

    df = pd.DataFrame(rows).sort_values('error_rate', ascending=False)
    print(f'\n{m["name"]} — Per-class error rate (top-{top_n} worst):')
    print(df.head(top_n).to_string(index=False))
    return df

err_b = print_per_class_error_rate(metrics_b, results_b)
err_f = print_per_class_error_rate(metrics_f, results_f)
````

````python
# --- 6.3 Фреквентные пары ошибок GT -> Pred ---

def top_error_pairs(m: dict, res: dict, n: int = 20):
    gt_v   = m['gt_v']
    top1_v = m['top1_v']
    labels = res['labels']
    wrong  = gt_v != top1_v
    pairs  = list(zip(gt_v[wrong].tolist(), top1_v[wrong].tolist()))
    pair_counts = {}
    for g, p in pairs:
        k = (labels[g], labels[p])
        pair_counts[k] = pair_counts.get(k, 0) + 1
    top = sorted(pair_counts.items(), key=lambda x: -x[1])[:n]
    print(f'\n{m["name"]} — Top-{n} confusion pairs (GT → Predicted):')
    for (gt_l, pred_l), cnt in top:
        print(f'  {gt_l:20s} -> {pred_l:20s}: {cnt}')

top_error_pairs(metrics_b, results_b)
top_error_pairs(metrics_f, results_f)
````
